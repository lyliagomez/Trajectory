Ploting trajectory of object with and without drag.

```python

from matplotlib import pyplot as plt
import math

xdrag=[0]
ydrag=[2]
g= -9.81
dt= 0.0001
x=0
y=2
v0=15
th=3.14/4
m=0.6
d=0.24
vx= v0*math.cos(th)
vy= v0*math.sin(th)
c= 0.25*d**2/m
b= d*1.6e-4/m

while (y>0):
  dx=vx*dt
  dy=vy*dt
  x =x+dx
  y =y+dy
  v=math.sqrt(vx**2 + vy**2)
  ax= -c*v*vx - b*vx
  ay= -c*v*vy - b*vy +g
  dvx=ax*dt
  dvy=ay*dt
  vx=vx+dvx
  vy=vy+dvy
  xdrag.append(x)
  ydrag.append(y)
plt.scatter(xdrag,ydrag)
plt.xlim(0,20)
plt.ylim(0,7.5)
```
